```python
import pandas as pd
import numpy as np
df = pd.read_csv('Sample - Superstore.csv')
df.head()

df.info()

df.describe()
df.isnull().sum()
df.duplicated().sum()
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Year'] = df['Order Date'].dt.year

df['Month'] = df['Order Date'].dt.month_name()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
ANALYSIS 1
SALES BY CATEGORY
df.groupby('Category')['Sales'].sum()
ANALYSIS 2
PROFIT BY CATEGORY
df.groupby('Category')['Profit'].sum()
BUSINESS INSIGHT: High sales does not always mean high profit.
ANALYSIS 3
TOP 10 PRODUCTS
top_products = df.groupby('Product Name')['Profit'].sum()

top_products.sort_values(ascending=False).head(10)
ANALYSIS 4
LOSS MAKING PRODUCTS
top_products.sort_values().head(10)
ANALYSIS 5
REGION PERFORMANCE
df.groupby('Region')[['Sales','Profit']].sum()
BAR CHART:
region_analysis = df.groupby('Region')[['Sales', 'Profit']].sum().reset_index()

print(region_analysis)
SALES BY REGION BAR CHART
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.bar(region_analysis['Region'],
        region_analysis['Sales'])

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')

plt.show()
PROFIT BY REGION BAR CHART
plt.figure(figsize=(8,5))

plt.bar(region_analysis['Region'],
        region_analysis['Profit'])

plt.title('Profit by Region')
plt.xlabel('Region')
plt.ylabel('Total Profit')

plt.show()
BETTER LOOKING CHART BY SEABORN
import seaborn as sns

plt.figure(figsize=(8,5))

sns.barplot(
    data=region_analysis,
    x='Region',
    y='Sales'
)

plt.title('Total Sales by Region')

plt.show()
COMBINED SALES AND PROFIT CHART
region_analysis.plot(
    x='Region',
    y=['Sales','Profit'],
    kind='bar',
    figsize=(10,6)
)

plt.title('Sales and Profit by Region')
plt.ylabel('Amount')

plt.show()
ANALYSIS 6
CUSTOMER SEGMENT ANALYSIS
df.groupby('Segment')[['Sales','Profit']].sum()
ANALYSIS 7
MONTHLY SALES TREND
monthly = df.groupby('Month')['Sales'].sum()
```
